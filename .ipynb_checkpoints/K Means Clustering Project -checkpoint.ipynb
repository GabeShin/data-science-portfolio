## K Means Cluster Creation
_______

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2)
kmeans.fit(df.drop('Private', axis =1))
```

** What are the cluster center vectors?**

```python
kmeans.cluster_centers_
```

## Evaluation
_________


Normally, there's no way to evaluate clustering if you don't have the labels. 

However, for this exercise, we do have the labels, let's evaluate the model.

```python
def is_it_private(input):
    if input == 'Yes':
        return 1
    else:
        return 0
df['Original'] = df['Private'].apply(is_it_private)
```

```python
df.sample(3)
```

** Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.**

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Original'],kmeans.labels_))
print(classification_report(df['Original'],kmeans.labels_))
```

## Conclusion

- Not a bad prediction! The algorithm only uses numeric features to cluster universities into 2 distinct group and I am happy with 76% precision rate.
